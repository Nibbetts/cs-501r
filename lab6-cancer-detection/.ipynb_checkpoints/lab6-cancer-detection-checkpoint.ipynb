```python
# Nathan Tibbetts
# Lab 6 - Cancer Detection

%matplotlib notebook

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
from os import listdir
from os.path import isfile, join
#from IPython.display import display, Image

TRAIN_PATH = "cancer_data/train"
TEST_PATH = "cancer_data/test"
# Load Data:
train_in_files  = [f for f in listdir(join(TRAIN_PATH, "inputs"))  if isfile(join(join(TRAIN_PATH, "inputs"),  f))]
train_out_files = [f for f in listdir(join(TRAIN_PATH, "outputs")) if isfile(join(join(TRAIN_PATH, "outputs"), f))]
test_in_files   = [f for f in listdir(join(TEST_PATH, "inputs"))   if isfile(join(join(TEST_PATH, "inputs"),   f))]
test_out_files  = [f for f in listdir(join(TEST_PATH, "outputs"))  if isfile(join(join(TEST_PATH, "outputs"),  f))]

assert (train_in_files == train_out_files), "Mismatch in direcory contents between train-in and train-out files!"
assert (test_in_files == test_out_files), "Mismatch in directory contents between test-in and test-out files!"
N = len(train_in_files) # Number of Test images to use.
T = len(test_in_files)  # Number of Train images to use.
S = 10   # Mini-batch size - shared for both train and test.
Ri = 256 # Resolution to reduce to before starting
Rf = 64  # Resolution to reduce to by the end
IN_SHAPE  = (Ri, Ri, 3) # Used only to scale down the images we work with - change for final trainings!
OUT_SHAPE = (Rf, Rf)    # Match back end to front end, but only one channel.
BATCH_IN_SHAPE  = (S, Ri, Ri, 3)
BATCH_OUT_SHAPE = (S, Rf, Rf)

train_neg = [f for f in train_in_files if f[:3] == "neg"]
train_pos = [f for f in train_in_files if f[:3] == "pos"]
test_neg  = [f for f in test_in_files  if f[:3] == "neg"]
test_pos  = [f for f in test_in_files  if f[:3] == "pos"]
assert (len(train_neg) + len(train_pos) == len(train_in_files)), "Not all train files identifiable as 'pos' or 'neg'!"
assert (len(test_neg)  + len(test_pos)  == len(test_in_files)),  "Not all test files identifiable as 'pos' or 'neg'!"

#print "Accuracy Baseline is: %f" % ((len(train_pos)/float(len(train_in_files)) * len(test_pos) +
#                                    len(train_neg)/float(len(train_in_files)) * len(test_neg)) / len(test_in_files))
#     % we say pos. * num real pos. cases + % we say neg. * num real neg. cases, all / by num. cases to make a %.
 
example_index = test_in_files.index("pos_test_000072.png")
    
features      = np.array([resize(imread(join(TRAIN_PATH, "inputs",  f)), IN_SHAPE)
                          for f in train_in_files[:N]],  dtype='float32')
labels        = np.array([resize(imread(join(TRAIN_PATH, "outputs", f)), OUT_SHAPE)
                          for f in train_in_files[:N]], dtype='int32')
test_features = np.array([resize(imread(join(TEST_PATH,  "inputs",  f)), IN_SHAPE)
                          for f in test_in_files[:T]],   dtype='float32')
test_labels   = np.array([resize(imread(join(TEST_PATH,  "outputs", f)), OUT_SHAPE)
                          for f in test_in_files[:T]],  dtype='int32')
# Note: Above we're not using train_out_files or test_out_files because we have asserted they are the same,
#       and if we shuffle them we want them to still be paired.

# plt.imshow(labels[0])

# Whiten our data - zero mean and unit standard deviation
mean = np.mean(features)#, axis=0) # I think the best way to put these on the same page is to scale them the same,
std  = np.std(features)#, axis=0)  #     which is based on the scaling for our features we know about beforehand.
features      = (features      - mean) / std # REMOVED axis=0 ... MAKES SENSE ???
test_features = (test_features - mean) / std
# features = np.transpose(np.reshape(features,(N,3,32,32)),axes=[0,2,3,1])
# features is (N, 1024, 1024, 3), or (batch_size, height, width, channels)

#plt.imshow(features[N-1]*0.1 + 0.5)

HEIGHT = np.shape(features)[1]
WIDTH = np.shape(features)[2]
CHANNELS = np.shape(features)[3]
```

```python
# Layer Defining helper functions

# Convolution Layer:
def conv( x, filter_size=3, stride=2, num_filters=64, is_output=False, name="conv" ):
    '''
    x is an input tensor
    Declare a name scope using the "name" parameter
    Within that scope:
      Create a W filter variable with the proper size
      Create a B bias variable with the proper size
      Convolve x with W by calling the tf.nn.conv2d function
      Add the bias
      If is_output is False,
        Call the tf.nn.relu function
      Return the final op
    ''' 
    with tf.variable_scope(name):
        W = tf.get_variable("W",
                            shape=(filter_size,
                                   filter_size,
                                   x.get_shape().as_list()[3],
                                   num_filters),
                            initializer=tf.contrib.layers.variance_scaling_initializer())
        b = tf.get_variable("b",
                            shape=num_filters,
                            initializer=tf.contrib.layers.variance_scaling_initializer())
        # Notice the input x will be one minibatch at a time,
        #    but right now we're just building a computation graph!
        h = tf.nn.conv2d(x, W, [1,stride,stride,1], padding="SAME", name="h")
        h = tf.nn.bias_add(h, b, name="h")
        if not is_output:
            h = tf.nn.relu(h, name="h_hat")
        l1 = tf.reduce_sum(W) + tf.reduce_sum(b)
        l2 = tf.reduce_sum(tf.square(W)) + tf.reduce_sum(tf.square(b))
        return h, l1, l2

# Fully-Connected Layer:
def fc( x, out_size=50, is_output=False, name="fc" ):
    '''
    x is an input tensor
    Declare a name scope using the "name" parameter
    Within that scope:
      Create a W filter variable with the proper size
      Create a B bias variable with the proper size
      Multiply x by W and add b
      If is_output is False,
        Call the tf.nn.relu function
      Return the final op
    '''
    with tf.variable_scope(name):
        W = tf.get_variable("W",
                            shape=(out_size,x.get_shape().as_list()[0]),
                            initializer=tf.contrib.layers.variance_scaling_initializer())
        b = tf.get_variable("b",
                            shape=(out_size,1),#out_size,
                            initializer=tf.contrib.layers.variance_scaling_initializer())
        h = tf.matmul(W, x, name="h") + b
        #h = tf.nn.bias_add(h, b, name="h")#(tf.reshape(h,[-1]), b, name="h")
        if not is_output:
            h = tf.nn.relu(h, name="h_hat")
        l1 = tf.reduce_sum(W) + tf.reduce_sum(b)
        l2 = tf.reduce_sum(tf.square(W)) + tf.reduce_sum(tf.square(b))
        return h, l1, l2
```

```python
# Setting up the Computation Graph

# Parameters:
c = 0.001 # Learning Rate
lam = 0.01 # up to 0.05
# Regularization choices:
l1_reg = False
l2_reg = False
dropout = False

tf.reset_default_graph()

x      = tf.placeholder(dtype=np.float32, shape=BATCH_IN_SHAPE,  name='x'     )
y      = tf.placeholder(dtype=np.int32,   shape=BATCH_OUT_SHAPE, name='y'     )
x_test = tf.placeholder(dtype=np.float32, shape=BATCH_IN_SHAPE,  name="x_test")
y_test = tf.placeholder(dtype=np.int32,   shape=BATCH_OUT_SHAPE, name="y_test")
training = tf.placeholder(dtype=np.bool, shape=(), name="training")

n = 4 # Number of Hidden Layers
l1 = [0]*n
l2 = [0]*n

"""
    Topology:
        My network topology is not very complicated, but it doesn't seem to do much better
    than when it was a single convolution layer. Instead I have three convolution layers,
    with dropout in between, and relu activations on the first two.
        Note: I chose to allow the end result to remain a lower resolution than the input,
    partially for lack of time, and partially since the simplicity of the white regions does not require
    a highly detailed view. Also, I did not know how to up-convolve, and ran out of time to ask.
"""
h0, l1[0], l2[0] = conv(x,  stride=1, name='conv0', num_filters=32,  is_output=False) # h0.shape = [S,256,256,32]
if dropout: h0 = tf.layers.dropout(h0, training=training, name='dropout0')
h1, l1[1], l2[1] = conv(h0, stride=2, name='conv1', num_filters=83,  is_output=False) # h1.shape = [S,128,128,83]
if dropout: h1 = tf.layers.dropout(h1, training=training, name='dropout1')
h2, l1[2], l2[2] = conv(h1, stride=2, name='conv2', num_filters=2,  is_output=True ) # h2.shape = [S,64,64,2]
hn = h2

"""
    Notes: While I did not end up with a more complicated network here, I did attempt a couple more layers,
    with poor results, and I did attempt something very similar to a tiny U-network in the last lab,
    before I knew how, and also did not get good results then - probably because I had fully connected layers
    in the wrong places.
"""
# shape1 = h0.get_shape().as_list()#h6.get_shape().as_list()
# flat1 = tf.transpose(tf.reshape(h6,[shape1[0],shape1[1]*shape1[2]*shape1[3]]))
# flat1 is [512,S]
# fc1 = fc(flat1,out_size=64,name='fc1')# fc1 is [64,S]
# fc2 = fc(fc1,out_size=10,name='fc2',is_output=True) # fc2 is [10,S]

"""
    Cost Function:
        My cost function is also simple and straightforward, while still allowing
    for either L1 or L2 regularization. It uses cross-entropy.
"""
with tf.name_scope(name="train_loss") as scope:
    xent = tf.nn.sparse_softmax_cross_entropy_with_logits(
        labels=y,
        logits=hn,
        name='cross_entropy')
    l1regularizer = l1_reg * tf.cast(training, tf.float32) * tf.reduce_sum(l1)          * lam
    l2regularizer = l2_reg * tf.cast(training, tf.float32) * tf.sqrt(tf.reduce_sum(l2)) * lam
    loss = tf.reduce_mean(xent) + l1regularizer + l2regularizer

"""
    Accuracy:
        Method: mean of pixels correct. This could have probably been improved (see description below of
    baseline accuracy.)
"""
with tf.name_scope(name="accuracy") as scope:
    prediction = tf.cast(tf.argmax(hn,axis=-1),tf.int32)
    correct_prediction = tf.equal(prediction,y)
    accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))

# Training Optimizer:
train_step = tf.train.AdamOptimizer(learning_rate=c,name='adam_optimizer').minimize(xent)

# Summaries:
with tf.variable_scope("train_summary"):
    tf.summary.scalar('xent_loss', loss)
    tf.summary.scalar('accuracy', accuracy)
    train_merged = tf.summary.merge_all()

with tf.variable_scope("test_summary"):
    tf.summary.scalar('xent_loss', loss)
    tf.summary.scalar('accuracy', accuracy)
    test_merged = tf.summary.merge_all()
```

```python
# Training

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    train_writer = tf.summary.FileWriter("./tf_logs", sess.graph)
    
    """
        Baseline Accuracy
            Here I have attempted to take a measurement of what the baseline for my accuracy would be,
        but I noticed over multiple runs it varied drastically irrespective of the number of batches
        we run it on (num). I don't trust it, but I assume generally that this value should be around 85,
        because there are many more dark pixels than light overall.
            Perhaps a better measure of accuracy overall would have been how many images we flagged, period,
        or add that to the loss somehow. ie: my 91% test accuracy is really not very good at all.
            I had first tried to predict this statistically, near the very top of my code, commented out,
        based on ratios, but to no avail. Everything method which seemed conceptually correct came out to 0.5,
        and I gave up on anything but brute force or observation.
    """
    num = 50
    base_acc = [0]*num
    for t in xrange(num):
        ix = np.random.randint(T,size=S)
        base_acc[t] = sess.run(accuracy, feed_dict={x:test_features[ix], y:test_labels[ix], training:False})
    print "Baseline Accuracy: %f" % np.mean(base_acc)
    print

    for t in xrange(750):#(N/S):
    #for t in np.ones(50,dtype=np.int32): # Overfitting test
        ix = np.random.randint(N,size=S)
        _, loss_value, acc, graph_summary = sess.run(
            [train_step, loss, accuracy, train_merged],
            #feed_dict={x:features[S*t:S*(t+1)],y:labels[S*t:S*(t+1)]})
            feed_dict={x:features[ix], y:labels[ix], training:True})
        train_writer.add_summary(graph_summary, t)
        if t % S == 0:
            ix = np.random.randint(T,size=S)
            tt = t / S
            test_loss_value, test_acc, test_summary = sess.run(
                [loss, accuracy, test_merged],
                #feed_dict={x:test_features[S*tt:S*(tt+1)], y:test_labels[S*tt:S*(tt+1)]})
                feed_dict={x:test_features[ix], y:test_labels[ix], training:False})
            train_writer.add_summary(test_summary, t)
            print "\t", test_loss_value, test_acc
        print t, loss_value, acc
        
    train_writer.close()
    
    # Show off an example
    example = sess.run(prediction, feed_dict={
        x:np.concatenate(([test_features[example_index]],
                          np.zeros((S-1,HEIGHT,WIDTH,CHANNELS),dtype=np.float32))),
        training:False})[0]
    #example = resize(example, (1024, 1024))
    plt.imshow(example, cmap="gray")
    
    """
        pos_test_000072.png
            Voila, as requested (at the very bottom of this round of outputs).
            I decided to leave this in a lower resolution since I did not actually do computations heavy enough
        to produce valid results at higher resolution - though I did once check with a brief calculation,
        I did not think to save it. (Oops.)
    """
```
